```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

# Step 1: Load data

```python
(train_data, train_labels), (test_data, test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
X_val = train_data[-10_000:]
y_val = train_labels[-10_000:]

X_train = train_data[:-10_000]
y_train = train_labels[:-10_000]
X_test = test_data
y_test = test_labels
```

```python
X_train.shape, X_test.shape, X_val.shape
```

```python
X_train[0][5]
```

```python
X_train = X_train / 255
X_val = X_val / 255
X_test = X_test / 255
```

```python
X_train[0][5]
```

```python
X_train.shape
```

```python
X_train_reshape = X_train.reshape(50_000,28*28)
X_val_reshape = X_val.reshape(10_000, 28*28)
X_test_reshape = X_test.reshape(10_000, 28*28)
```

```python
X_train_reshape.shape, X_val_reshape.shape, X_train_reshape.shape
```

```python
y_train.shape
```

```python
y_train[0]
```

# Step 2: Build a MLP

```python
def get_model():
  model = tf.keras.Sequential([
      tf.keras.Input(shape=(784,)),
      tf.keras.layers.Dense(units=300, activation="relu"),
      tf.keras.layers.Dense(units=100, activation="relu"),
      tf.keras.layers.Dense(units=10, activation="softmax")
  ])

  return model
```

```python
model = get_model()
model.summary()
```

# Step 3: Compile Model

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

# Step 4: Train model

```python
earlystop_cb = tf.keras.callbacks.EarlyStopping(patience=5)
model_checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(filepath=".")

model.fit(X_train_reshape, y_train, validation_data=(X_val_reshape, y_val), batch_size=256, epochs=30, callbacks=[earlystop_cb, model_checkpoint_cb])
```

```python
model.evaluate(X_test_reshape, y_test)
```
