```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(z))

xs = np.linspace(-5, 5, 1200)
plt.plot(xs, sigmoid(xs))
```

```python
def relu(x):
    return np.maximum(0, x)
# of np.where(x > 0, x, 0)

plt.plot(xs, relu(xs))
```

```python
def Tanh(x):
    # return np.tanh(x)
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(xs, Tanh(xs))
```

```python
def step(x):
    return np.where(x > 0, 1, 0)

plt.plot(xs, step(xs))
```
