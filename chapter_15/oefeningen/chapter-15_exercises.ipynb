# Jena Climate Data

```python
import tensorflow as tf
import shutil
import pandas as pd
import numpy as np
```

## Step 1

```python
data = tf.keras.utils.get_file(origin="https://s3.amazonaws.com/keras-datasets/jena_climate_2009_2016.csv.zip")
```

```python
data_unzip = shutil.unpack_archive(data)
```

```python
df = pd.read_csv("./jena_climate_2009_2016.csv")
df.head()
```

```python
temperature = np.array(df['T (degC)'])
raw_data = np.array(df.drop(['Date Time'], axis=1))
```

```python
df.shape
```

```python
temperature.shape, raw_data.shape
```

## Step 2

```python
perc_50 = df.shape[0] // 2
perc_25 = df.shape[0] // 4
```

```python
X_train = raw_data[:perc_50]
y_train = temperature[:perc_50]

X_train.shape, y_train.shape
```

```python
X_test = raw_data[-perc_25:]
y_test = temperature[-perc_25:]

X_test.shape, y_test.shape
```

```python
X_val = raw_data[perc_50: -perc_25]
y_val = temperature[perc_50: -perc_25]

X_val.shape, y_val.shape
```

### Normalisatie

```python
mu = np.mean(X_train, axis=0)
sigma = np.std(X_train, axis=0)
```

```python
X_train_norm = ((X_train - mu) / sigma)
X_test_norm = ((X_test - mu) / sigma)
X_val_norm = ((X_val - mu) / sigma)
```

```python
X_train[0][1]
```

```python
X_train_norm[0][1]
```

```python
(X_train_norm[0][1] * sigma[1]) + mu[1]
```

# create dataset

```python
def create_ds(data, label, train=False, bs=256):
  ds = tf.keras.utils.timeseries_dataset_from_array(
      data=data,
      targets=label[(120 * 6 + 23 * 6):],
      sequence_length=120,
      sampling_rate=6,
      batch_size=bs,
      shuffle=train
      )
  return ds
```

```python
train_ds = create_ds(X_train_norm, y_train, train=True)
test_ds = create_ds(X_test_norm, y_test, train=False)
val_ds = create_ds(X_val_norm, y_test, train=False)
```

```python
my_check_ds = create_ds(
    data=df.values[:1000, 0],
    label=df.values[:1000, 0],
    train=False,
    bs=1
)

for features, target in my_check_ds.take(2):
  print(f"{features} maps to {target}")
  print("*"*50)
```

# Step 3: Establish a Common Sense Baseline

```python
def compute_naive_mae(dataset):
  total = 0
  for f, l in dataset:
    f = (f * sigma) + mu
    for idx, i in enumerate(f):
      total += abs(i[:, 1] - l[idx])
  return total/dataset.cardinality().numpy()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation="relu", input_shape=(120, 14)),
    tf.keras.layers.Dense(14, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(), metrics=["mean_absolute_errror"], loss="sparse_categorical_crossentropy")
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath='/tmp/checkpoint', save_best_only=True)
```

```python
model.fit(train_ds, validation_data=val_ds, callbacks=[checkpoint], epochs=10)
```
